```python
df['Age']=df['age'].astype(int)
#throws error because null values present
```

```python
df['EnrollmentDate']= pd.to_datetime(df['EnrollmentDate'])
df.head(3)
```

```python
df.dtypes
```

```python
df['Age']
```

```python
#finding median for filling null values
df['Age'].median()
```

```python
df['Age']= df['Age'].fillna(df['Age'].median())#the null values of age is filled here so we can now convert age column into int type
```

veryifying that null values are cleared or not from age column

for score we use mean

```python
df['Score'].mean()
```

```python
df['Score']=df['Score'].fillna(df['Score'].mean())#since null values cleared we can convert age,score values into integer
```

```python
df['Age']=df['Age'].astype(int)
df['Score']=df['Score'].astype(int)
```

display the converted datatypes of age,score , enrollment

```python
df.isnull().sum()
```

```python
df['Grade'].mode()
#since we use mode here because we display most grade as common grade for students
```

```python
df['Email']=df['Email'].fillna('no email')
```

DATA VISUALIZATION


```python
import pandas as pd
path = '/content/project_dataset_unclean.csv'
df = pd.read_csv(path)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df
```

```python
df['Grade'] = df['Grade'].astype(str)
```

```python
df['Grade'].value_counts()
```

```python
# Now create the bar plot
sns.barplot(x='Grade', y='Age', data=df)
plt.title('count of students by grade with respective age')
plt.show()
```

```python
df['Passed'].value_counts()
```

PIECHART

```python
df['Passed'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('passed students')
plt.show()
```

HISTOGRAM

```python
df['Score'].hist(bins=10)
plt.title('score distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

BOX PLOT

```python
sns.boxplot(x='Grade',y='Score', data=df)
plt.title('score distribution')
plt.show()
```

SCATTERED PLOT

```python
plt.scatter(df['Age'], df['Score'])
plt.title('Age vs Score')
plt.xlabel('Age')
plt.ylabel('Score')
plt.show()
```
